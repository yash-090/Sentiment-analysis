```python
#making a GET request to the API
r=requests.get('https://newsapi.org/v2/top-headlines?country=in&category=science&apiKey=057fe98f8cf64453a0bb2b0d4f17440c')
r
```

```python
#loading the content with the help of json in a more readable format
data=json.loads(r.content)
data
```

```python
import pandas as pd
```

```python
#creating a dataframe
df=pd.DataFrame(data['articles'])
df
```

```python
#importing textblob library to perform sentiment analysis
from textblob import TextBlob
```

```python
import numpy as np
```

```python
#creating a function to check the polarity of the content of the article
def polarity(text):
    return TextBlob(text).sentiment.polarity
    
```

```python
#replacing the nonetypes in the dataframes with empty string values as textblob does not accept nonetypes
df.fillna("",inplace=True)
```

```python
df
```

```python
#calculating polarity on the content column
df['Polarity']= df['content'].apply(polarity)
df
```

```python
#creating a function to determine the sentiment on the basis of the polarity
def sentiment(value):
    if value < 0:
        return 'negative'
    elif value==0:
        return 'neutral'
    else:
        return 'positive'
    
```

```python
df['Sentiment']= df['Polarity'].apply(sentiment)
```
